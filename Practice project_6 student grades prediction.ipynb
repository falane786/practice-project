# encoded the data usign ordinalencoder

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_enc = OrdinalEncoder()
for i in df.columns:
    if df [i].dtypes=='object':
        df[i] = ord_enc.fit_transform(df[i].values.reshape(-1,1))
df
```

# split the features and target veriable

```python
X = df.drop(columns=['CGPA'])
y = df['CGPA']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# train the data

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(X)
```

```python
scaled
```

```python
X_train,X_test,y_train,y_test = train_test_split(scaled,y, test_size=0.27, random_state=56)
y_train.head()
```

# train the model

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

# train score

```python
regression.score(X_train, y_train)
```

# test score

```python
regression.score(X_test,y_test)
```

```python
X_test[1]
```

# predict the test data

```python
y_prediction=regression.predict(X_test)
```

```python
y_prediction
```

# Hyperparameter tuning using RandomizedSearchCv
